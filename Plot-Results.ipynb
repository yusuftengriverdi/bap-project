```python
tasks = [
    'fevsnonfe-below20mg',
    'multidensity-allvs200',
    'multiclass-b20mg-inc2',
    'multiclass-fe-density',
    'binarydensity-pb',
    'multiclass-200mg'
]
```

```python
BASE_MODELS = [
    'unet',
    'vgg5_block',
    'vgg16',
    'vgg19',
    'resnet50', 
    'deeplabv3plus',
    'nasnetmobile', 
    'inception-resnetv2', 
    'double_unet',
    'cnn_global_avg',
    'test_cnn_dense']

date_range = ['2022-03-28', '2022-03-30', '2022-03-31', '2022-04-01', '2022-04-03', '2022-04-04', '2022-04-07','2022-04-10', '2022-04-11']
img_size = 128
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
from labellines import labelLine, labelLines
```

```python
import warnings
warnings.filterwarnings('ignore')

#for task in tasks:
task = 'multiclass-fe-density'
fig, ax = plt.subplots()
fig.set_figheight(15)
fig.set_figwidth(15)
for BASE_MODEL in BASE_MODELS:
    logs = []
    for date in date_range:
        BASE_MODEL_NAME = f"{date}_{BASE_MODEL}_{img_size}_{task}"
        path = f'./logs/{task}/logs_{BASE_MODEL_NAME}.csv'
        if os.path.exists(path):  
            log = pd.read_csv(path)
            key = log.keys()[0]
            epochs = range(0, len(log['val_accuracy']))
            ax.set_ylim(0,1)
            ax.set_xlim(0,25)
            plt.plot(epochs, log[log.keys()[6]], '-', label=f'{BASE_MODEL}')
            # plt.plot(epochs, log[key], ':', label=f'{BASE_MODEL}')     
        else:
            #print(BASE_MODEL_NAME, 'cannot be found')
            pass
    plt.title(f'{key} Performance of Models')
    labelLines(ax.get_lines(), zorder=2.5, xvals=[1, 15, 6, 7, 8, 10, 12, 14, 17])

print(log.keys())
```

```python
fig, axes = plt.subplots(1, 2)
fig.set_figheight(20)
fig.set_figwidth(20)

axes = axes.flatten()
for i, BASE_MODEL in enumerate(BASE_MODELS):
    logs = []
    ax = axes[i]
    for date in date_range:
        BASE_MODEL_NAME = f"{date}_{BASE_MODEL}_{img_size}_{task}"
        path = f'./logs/{task}/logs_{BASE_MODEL_NAME}.csv'
        if os.path.exists(path):  
            log = pd.read_csv(path)
            logs.append(log)
            epochs = range(0 ,len(log['accuracy']))
            #.set_ylim(0,1)
            ax.plot(epochs, log['loss'], '-.', label=f'train_loss')
            ax.plot(epochs, log['val_loss'], '-', label=f'valid_loss')  
            ax.set_title(f'{BASE_MODEL}')
        else:
            pass
            #print(BASE_MODEL_NAME, 'cannot be found')
    labelLines(ax.get_lines(), align='False')
```
