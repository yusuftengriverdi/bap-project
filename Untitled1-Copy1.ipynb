```python
import tensorflow as tf
import keras
import tensorflow_addons as tfa
import numpy as np
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
#import seaborn as sns
from scipy.ndimage import rotate
import random
import cv2
import os
from tqdm import tqdm
from tensorflow.keras import datasets, layers, models
from keras.preprocessing.image import ImageDataGenerator
from keras.applications import VGG16, ResNet50

import pandas as pd

from pickle import dump
from keras.models import Model
```

```python
def stack_data(number_of_samples= 1000, directory= "data\\dots\\", im_size = 256, validation_split=0.8):

    folders = os.listdir(directory)
    #features = ['dot']
    
    file_list = []
    classes = ['2', '3', '4']
    one_hot_vector = np.eye(len(classes))

    tag= 0
    for folder in folders:
        counter = 0
        print(folder, one_hot_vector[tag])
        image_names = os.listdir(os.path.join(directory, folder))
        for name in tqdm(image_names):
            if (counter == number_of_samples):
                break
            try:
                img_list = []
                img = cv2.imread(os.path.join(directory, folder, name))
                #print(os.path.join(directory, folder, name))
                img = cv2.resize(img, (im_size,im_size), cv2.INTER_CUBIC)
                img_list.append(img)
                class_label = one_hot_vector[tag]
                img_list.append(class_label)
                file_list.append(np.asarray(img_list))
                counter +=1
            except Exception as e:
                pass#print(e)
        tag +=1

    data = np.asarray(file_list) 
    np.random.shuffle(data)
    print(len(data))

    boundary = int(len(data) * 0.8)
    
    train_set = data[:boundary]
    train_images = np.asarray([i[0] for i in train_set] ) 
    train_labels = np.asarray([i[1] for i in train_set] )

    valid_set = data[boundary:(len(data)-1)]
    valid_images = np.asarray([i[0] for i in valid_set] ) 
    valid_labels = np.asarray([i[1] for i in valid_set] )


    test_set = data[len(data)-1:]
    test_images = np.asarray([i[0] for i in test_set] ) 
    test_labels = np.asarray([i[1] for i in test_set] )

    print("train", train_images.shape, train_labels.shape)
    print("valid",valid_images.shape, valid_labels.shape)
    print("test",test_images.shape, test_labels.shape)

    return train_images, train_labels, valid_images, valid_labels, test_images, test_labels
```

```python
 train_images, train_labels, valid_images, valid_labels, test_images, test_labels  = stack_data()
```

```python
#KERAS PREPROCESSING 
train_images = tf.keras.applications.resnet.preprocess_input(train_images)

valid_images = tf.keras.applications.resnet.preprocess_input(valid_images)

test_images = tf.keras.applications.resnet.preprocess_input(test_images)
```

```python
convNet = ResNet50(weights= None, input_shape = (256,256,3), classes=3)
convNet.summary()
```

```python
optimizer = keras.optimizers.SGD(lr=0.001, momentum=0.9)

convNet.compile(optimizer=optimizer,
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy',  tfa.metrics.F1Score(num_classes=3, threshold=0.5)])
```

```python
# save model
# MODEL NAME ANNOTATIONS: MODELTYPE_ANYMODEL_CHANGE_OPTIMIZERTYPE_LR_MOMENTUM_LOSSFN_DATAKIND_NUMOFSAMPLES_VALIDATIONSPLIT_NUMCLASSES_INPUTSIZE_CHANELNUMBERS_ANYSPECIFICPREPROCESSING_BATCHSIZE_EPOCHS
MODEL_NAME = "Resnet50_SGD_1e-3_09_catce_2mg_18_0.8_3_256_3_keras_3_100"
```

```python
history = convNet.fit(train_images, train_labels, epochs=100, 
                      validation_data=(valid_images, valid_labels), batch_size=3)
```

```python

def plot_metrics(history, metric_name= 'accuracy'):
    plt.plot(history.history[metric_name], label=metric_name)
    plt.plot(history.history[f'val_{metric_name}'], label = f'val_{metric_name}')
    plt.xlabel('Epoch')
    plt.ylabel('Metric')
    plt.ylim([0.5, 1])
    plt.legend(loc='lower right')
    
plot_metrics(history)
```

```python
print(test_labels)
convNet.evaluate(test_images.reshape(1,256,256,3),  test_labels, verbose=2)
```

```python
#save history
def save_logs(history= history, model_name = MODEL_NAME):
    pd.DataFrame.from_dict(history.history).to_csv(f'history_{MODEL_NAME}.csv',index=False)
    df = pd.DataFrame.from_dict(history.history)
    return df.style.set_caption(f"{model_name}")

df = save_logs()

convNet.save(f'{MODEL_NAME}.h5')

df
```

```python
def VGG3(input_size= 256):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(input_size, input_size, 3)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu', kernel_initializer='he_uniform'))
    model.add(layers.Dense(3, activation='sigmoid'))
    # compile model
    opt = keras.optimizers.SGD(lr=0.00001, momentum=0.9)
    model.compile(optimizer=opt, loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True), metrics=['accuracy', tfa.metrics.F1Score(num_classes=3, threshold=0.5)])
    #print(model.summary())
    return model

net = VGG3()

history_v2 = net.fit(train_images, train_labels, epochs=100, validation_data=(valid_images, valid_labels),  
                     batch_size=3)

# MODEL NAME ANNOTATIONS: MODELTYPE_ANYMODEL_CHANGE_OPTIMIZERTYPE_LR_MOMENTUM_LOSSFN_DATAKIND_NUMOFSAMPLES_VALIDATIONSPLIT_NUMCLASSES_INPUTSIZE_CHANELNUMBERS_ANYSPECIFICPREPROCESSING_BATCHSIZE_EPOCHS
MODEL_NAME = "VGG3_SGD_1e-2_09_catce_2mg_18_08_3_256_3_keras_3_100"

 
```
